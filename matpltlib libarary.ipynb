## Line plot

```python

emp_age=[25,26,27,28]
emp_salary=[12000,3000,500000,16000]

plt.title('Salary for each age')
plt.xlabel('Age')
plt.ylabel('Salary')
mp_age=[25,26,27,28]
mp_salary=[12050,3080,50050,161000]

plt.plot(mp_age,mp_salary)

plt.plot(emp_age,emp_salary)


plt.legend(['doctors','enfbcvg'])
plt.savefig('salary')
plt.grid()
```

# Bar Chart

```python
emp_age=[25,26,27,28]
emp_salary=[120020,305500,510000,16610]

emp_indecies=np.arange(len(emp_age))

width=0.25

plt.title('Salary for each age')
plt.xlabel('Age')
plt.ylabel('Salary')

mp_age=[25,26,27,28]
mp_salary=[120230,308550,500650,161000]

plt.bar(emp_indecies-width/2,mp_salary,width=width)

plt.bar(emp_indecies+width/2,emp_salary,width=width)

plt.xticks(ticks=emp_indecies,labels=emp_age)
plt.legend(['doctors','enfbcvg'])
```

## Histograms


```python
age=[45,26,51,40,12,80,62,18,90,70]

mean=np.mean(age)
print(mean)

plt.hist(age,bins=7,edgecolor='black')
plt.axvline(mean,color='red')

plt.title('age dis')
plt.xlabel('age',color='r')
plt.ylabel('fequecy',color='r')
```

## Scatter Plots

```python
x=[1,2,3,4,6,9,7,5,4,5,3]
y=[2,3,6,5,2,6,9,7,8,0,7]
colors=[5,6,1,4,6,4,6,3,6,3,6]

plt.scatter(x,y,c=colors,s=100,edgecolor='black',linewidth=2,alpha=0.5,cmap='plasma')

c_bar=plt.colorbar()
c_bar.set_label('Colors bar')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('scatter plot test')

plt.show()
```
